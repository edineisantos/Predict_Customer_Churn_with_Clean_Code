```python
# Import libraries
import pandas as pd
from constants import file_path
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
def import_data(file_path):
    '''
    returns dataframe for the csv found at pth

    input:
            pth: a path to the csv
    output:
            dataframe: pandas dataframe
    '''

    dataframe = pd.read_csv(file_path)
    return dataframe
```

```python
churn_df = import_data(file_path)
```

```python
cat_columns = [
    'Gender',
    'Education_Level',
    'Marital_Status',
    'Income_Category',
    'Card_Category'                
]

quant_columns = [
    'Customer_Age',
    'Dependent_count', 
    'Months_on_book',
    'Total_Relationship_Count', 
    'Months_Inactive_12_mon',
    'Contacts_Count_12_mon', 
    'Credit_Limit', 
    'Total_Revolving_Bal',
    'Avg_Open_To_Buy', 
    'Total_Amt_Chng_Q4_Q1', 
    'Total_Trans_Amt',
    'Total_Trans_Ct', 
    'Total_Ct_Chng_Q4_Q1', 
    'Avg_Utilization_Ratio'
]
```

```python
churn_df['Churn'] = churn_df['Attrition_Flag'].apply(lambda val: 0 if val == "Existing Customer" else 1)
```

```python
# churn_distribution.png
plt.figure(figsize=(20,10)) 
churn_df['Churn'].hist();
```

```python
# customer_age_distribution.png
plt.figure(figsize=(20,10)) 
churn_df['Customer_Age'].hist();
```

```python
# heatmap.png
plt.figure(figsize=(20,10)) 
sns.heatmap(churn_df.corr(), annot=False, cmap='Dark2_r', linewidths = 2)
plt.show()
```

```python
# marital_status_distribution.png
plt.figure(figsize=(20,10)) 
churn_df.Marital_Status.value_counts('normalize').plot(kind='bar');
```

```python
# total_transaction_distribution.png
plt.figure(figsize=(20,10)) 
# distplot is deprecated. Use histplot instead
# sns.distplot(df['Total_Trans_Ct']);
# Show distributions of 'Total_Trans_Ct' and add a smooth curve obtained using a kernel density estimate
sns.histplot(churn_df['Total_Trans_Ct'], stat='density', kde=True);
```
